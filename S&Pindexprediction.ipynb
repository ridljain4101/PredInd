```python
##prediction --whether it goes down or up tomorrow
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500["Target"]= (sp500["Tomorrow"]>sp500["Close"]).astype(int)
```

```python
## a lot of market shifted from the time started for the index, we only take last 30 yrs data.
sp500=sp500.loc["1990-01-01":].copy()
```

```python
##prediction modelling

##RF--> lesser overfit, better non -linear catch.
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=1)
train=sp500.iloc[:-100]
test=sp500.iloc[-100:]

predictors=["Close","Volume","Open", "High","Low"]
model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score
preds=model.predict(test[predictors])
```

```python
precision_score(test["Target"],preds)
```

```python

def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict(test[predictors])
    preds=pd.Series(preds,index=test.index,name="Prediction")
    combined=pd.concat([test['Target'],preds],axis=1)
    return combined
```

```python
##back testing
def backtest(data,model,predictors,start=2500,step=250):
    all_predictions=[]

    for i in range(start,data.shape[0],step):
        train=data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions=predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)    
        
```

```python
predictions=backtest(sp500,model,predictors)
```

```python
predictions["Prediction"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Prediction"])
```

```python
predictions["Target"].value_counts()
```

```python
predictions["Target"].value_counts()/predictions.shape[0]

##the actual no of days it went up
```

```python
##additional predictors to improve it--rollinng avgs
##avvg cp in last 2 days, last week, last 3 months, last year, last 5 years
horizons=[2,5,60,250,1000]
new_predictors=[]
for horizon in horizons:
    rolling_avgs=sp500.rolling(horizon).mean()
    ratio_col=f"Close_Ratio_{horizon}"
    sp500[ratio_col]=sp500["Close"]/rolling_avgs["Close"]
    trend_col=f"Trend{horizon}"
    sp500[trend_col]=sp500.shift(1).rolling(horizon).sum()['Target']

    new_predictors+=[ratio_col,trend_col]

```

```python
##updating the model
model = RandomForestClassifier(n_estimators=200,min_samples_split=50,random_state=1)
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict_proba(test[predictors])[:,1]
    preds[preds>=0.6]=1
    preds[preds<0.6]=0
    preds=pd.Series(preds,index=test.index,name="Prediction")
    combined=pd.concat([test['Target'],preds],axis=1)
    return combined

##here wr find the probability that the price is going up or down, and if it is greater than 60, then we can say that
## it will surely go up.
##we only want trades where we can be much more sure that the price actually goes up, and we want to trade
## for lesser number of days.
```

```python
##backtesting-using ratio and trends which tell much more about the price going up or down
predictions=backtest(sp500,model,new_predictors)
```

```python
predictions["Prediction"].value_counts() 
```

```python
precision_score(predictions["Target"],predictions["Prediction"])
```
